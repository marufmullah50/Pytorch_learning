| **Data Type**             | **Dtype**         | **Description**                                                                                                                                                                |
|---------------------------|-------------------|--------------------------------------------------------------------------------------------------------------------------------------------------------------------------------|
| **32-bit Floating Point** | `torch.float32`   | Standard floating-point type used for most deep learning tasks. Provides a balance between precision and memory usage.                                                         |
| **64-bit Floating Point** | `torch.float64`   | Double-precision floating point. Useful for high-precision numerical tasks but uses more memory.                                                                               |
| **16-bit Floating Point** | `torch.float16`   | Half-precision floating point. Commonly used in mixed-precision training to reduce memory and computational overhead on modern GPUs.                                            |
| **BFloat16**              | `torch.bfloat16`  | Brain floating-point format with reduced precision compared to `float16`. Used in mixed-precision training, especially on TPUs.                                                |
| **8-bit Floating Point**  | `torch.float8`    | Ultra-low-precision floating point. Used for experimental applications and extreme memory-constrained environments (less common).                                               |
| **8-bit Integer**         | `torch.int8`      | 8-bit signed integer. Used for quantized models to save memory and computation in inference.                                                                                   |
| **16-bit Integer**        | `torch.int16`     | 16-bit signed integer. Useful for special numerical tasks requiring intermediate precision.                                                                                    |
| **32-bit Integer**        | `torch.int32`     | Standard signed integer type. Commonly used for indexing and general-purpose numerical tasks.                                                                                  |
| **64-bit Integer**        | `torch.int64`     | Long integer type. Often used for large indexing arrays or for tasks involving large numbers.                                                                                  |
| **8-bit Unsigned Integer**| `torch.uint8`     | 8-bit unsigned integer. Commonly used for image data (e.g., pixel values between 0 and 255).                                                                                    |
| **Boolean**               | `torch.bool`      | Boolean type, stores `True` or `False` values. Often used for masks in logical operations.                                                                                      |
| **Complex 64**            | `torch.complex64` | Complex number type with 32-bit real and 32-bit imaginary parts. Used for scientific and signal processing tasks.                                                               |
| **Complex 128**           | `torch.complex128`| Complex number type with 64-bit real and 64-bit imaginary parts. Offers higher precision but uses more memory.                                                                 |
| **Quantized Integer**     | `torch.qint8`     | Quantized signed 8-bit integer. Used in quantized models for efficient inference.                                                                                              |
| **Quantized Unsigned Integer** | `torch.quint8` | Quantized unsigned 8-bit integer. Often used for quantized tensors in image-related tasks.                                                                                     |

#2.Element wise Operation

#Element wise operation on single tensor

```python
c =torch.tensor([1,2,3,-8,9,-6])
```

```python
#absolute
abs1= torch.abs(c)
print(f"Absolute value of c tensor:\n {abs1}")
print("\n")

#Negative
neg1 = torch.neg(c)
print(f"Negative value of c tensor:\n {neg1}")
```

```python
d = torch.tensor([1.2,3.2,5.1,4.4])
```

```python
#round #closest round value
round1 = torch.round(d)
print(f"Round value of d tensor:\n {round1}")
print("\n")

#ceil #roundup upper value
ceil1 = torch.ceil(d)
print(f"Ceil value of d tensor:\n {ceil1}")
print("\n")

#floor #rounddown lower value
floor1 = torch.floor(d)
print(f"Floor value of d tensor:\n {floor1}")
print("\n")

#clamp #limit values within a range
clamp1 = torch.clamp(d, min=3, max = 4)
print(f"Clamp value of d tensor:\n {clamp1}")
```

#3.Reduction Operation

```python
e = torch.randint(size=(2,3), low=3 , high =10,dtype=torch.float32)
e
```

```python
#sum
sum1 = torch.sum(e)
#print(f"Sum of all elements in e tensor:\n {sum1}")
print("\n")
sum1
#sum along columns
colsum = torch.sum(e , dim=0)
print(f"Sum of elements along columns in e tensor:\n {colsum}")
print("\n")

#sum along rows
rowsum = torch.sum(e, dim=1)
print(f"Sum of elements along rows in e tensor:\n {rowsum}")
```

```python
sum1
```

```python
#mean
mean1 = torch.mean(e)
print(f"Mean of all elements in e tensor:\n {mean1}")
print("\n")

#mean along columns
colmean = torch.mean(e, dim=0)
print(f"Mean of elements along columns in e tensor:\n {colmean}")
print("\n")

#mean along rows
rowmean = torch.mean(e, dim=1)
print(f"Mean of elements along rows in e tensor:\n {rowmean}")
```

```python
#median
med1 = torch.median(e)
med1
```

```python
# max
max1 = torch.max(e)
print(f"Max value in e tensor:\n {max1}")
print("\n")

#min
min1 = torch.min(e)
print(f"Min value in e tensor:\n {min1}")
print("\n")

#priduct
prod1 = torch.prod(e)
print(f"Product of all elements in e tensor:\n {prod1}")
print("\n")

#Standard Deviation
std1 = torch.std(e)
print(f"Standard Deviation of all elements in e tensor:\n {std1}")
```

```python
#Argmax  
#Position of max value in tensor
argmax1 = torch.argmax(e)
print(f"Position of max value in e tensor:\n {argmax1}")
print("\n")

#Argmin
#Position of min value in tensor
argmin1 = torch.argmin(e)
print(f"Position of min value in e tensor:\n {argmin1}")
```

#6.Special function

```python
k = torch.randint(size=(2,3), low=2, high=10, dtype=torch.float32)
k
```

```python
#log
log1 = torch.log(k)
print(f"Logarithm of k tensor:\n {log1}")
print("\n")

# EXP
exp1 = torch.exp(k)
print(f"Exponential of k tensor:\n {exp1}")
print("\n")

#Square Root
sqrt1 = torch.sqrt(k)
print(f"Square Root of k tensor:\n {sqrt1}")
print("\n")

#Softmax
softmax1 = torch.softmax(k,dim=0)
print(f"Softmax of k tensor along dim 0:\n {softmax1}")
print("\n")

#relu
relu1 = torch.relu(k)
print(f"ReLU of k tensor:\n {relu1}")
```

# Copying Tensor

```python
v = torch.randint(size=(2,3), low=2, high=10)
v
```

```python
b = v
b
```

```python
v [0][0] = 0
v
```

```python
b 
# b changed as well which is not desirable because they share same memory location

print(f" id of v:{id(v)} \n id of b:{id(b)}")
```

```python
# So do clone to avoid this issue
s = v.clone()
s
```

```python
v[0][0] = 9
v
```

```python
s
```

```python
print(f" id of v:{id(v)} \n id of s:{id(s)}")
```
